```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Asus\Downloads\student_lifestyle_dataset.csv',encoding = 'unicode_escape')
```

```python
df.head()
```

```python
kpi_dict = {}
```

```python
kpi_dict['avg_study_hours'] = df['Study_Hours_Per_Day'].mean()
```

```python
kpi_dict['avg_sleep_hours'] = df['Sleep_Hours_Per_Day'].mean()

# KPI 3: Average extracurricular hours per day
kpi_dict['avg_extracurricular_hours'] = df['Extracurricular_Hours_Per_Day'].mean()

# KPI 4: Average social hours per day
kpi_dict['avg_social_hours'] = df['Social_Hours_Per_Day'].mean()

# KPI 5: Average physical activity hours per day
kpi_dict['avg_physical_activity_hours'] = df['Physical_Activity_Hours_Per_Day'].mean()

# KPI 6: Average GPA
kpi_dict['avg_GPA'] = df['GPA'].mean()
```

```python
stress_counts = df['Stress_Level'].value_counts()
```

```python
print('\
KPIs:')
for kpi, value in kpi_dict.items():
    print(kpi + ':', round(value, 2))

print('\
Stress Level Distribution:')
print(stress_counts)
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Study_Hours_Per_Day'], kde=True, color='green', bins=10)
plt.title('Distribution of Study Hours Per Day')
plt.xlabel('Study Hours Per Day')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['GPA'], kde=True, color='green', bins=10)
plt.title('GPA Distribution')
plt.xlabel('GPA')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Stress_Level', data=df, palette='Set2') 
plt.title('Stress Level Count')
plt.xlabel('Stress Level')
plt.ylabel('Count')
plt.show()
```

```python
recommendations = []

recommendations.append('1. If study hours are lower than optimal, consider implementing study support and time management workshops.')
recommendations.append('2. Monitor sleep hours: Encourage a balanced sleep schedule as insufficient sleep can affect academic performance.')
recommendations.append('3. Evaluate extracurricular and social activities: Too high a level of non-study activities might be affecting study time. Balance is key.')
recommendations.append('4. Physical activity plays a role in stress management. Promoting regular exercise could improve overall well-being and academic performance.')
recommendations.append('5. For students reporting high stress, provide counseling and stress management resources.')

print('\
Recommendations:')
for rec in recommendations:
    print(rec)

print('done')
```
